```python
def GetDataFrame(filename, columns = None):
    df = pd.read_csv(filename, encoding="utf-16", header=None)
    if columns is not None:
        df.columns = columns
    return df

def Agregate(dictionary):
    columns = ["date", "t_volume", "t_sum_pos", "t_sum_neg", "t_sum_neut", "t_sum_comp", "t_avg_neut", "t_avg_pos", "t_avg_neg", "t_avg_comp","t_zavg_neut", "t_zavg_pos", "t_zavg_neg", "t_zavg_comp", "t_nonZero_neut", "t_nonZero_pos", "t_nonZero_neg", "t_nonZero_comp", "t_day_of_week", "t_is_weekend"]
    df = pd.DataFrame(columns = columns)
    for key in dictionary:
        print(key +' is actually processing!', end='\r')
        df = ProcessDailyTwitterData(df, key)
    return df

def ProcessDailyTwitterData(df, filename):
    ddf = pd.read_csv(filename, encoding="utf-16", engine='python')
    row = []

    date = filename.split('/')[1][:-4]
    volume = len(ddf.index)
    #print("volume ",volume)
    
    nonZero_neut = (ddf.neutral>0.0).sum()
    #print("neutral ",nonZero_neut)
    
    nonZero_pos = (ddf.positive>0.0).sum()
    #print("positive ",nonZero_pos)
    
    nonZero_neg = (ddf.negative>0.0).sum()
    #print("negative ",nonZero_neg)
    
    nonZero_comp = (ddf['compound']>0.0).sum()
    #print("compound ",nonZero_comp)
    
    sum_neut = ddf.neutral.sum()
    sum_pos= ddf.positive.sum()
    sum_neg= ddf.negative.sum()
    sum_comp= ddf['compound'].sum()
    #print(sum_neut, sum_pos, sum_neg, sum_comp)
    
    avg_neut = sum_neut/volume
    avg_pos=  sum_pos/volume
    avg_neg=  sum_neg/volume
    avg_comp=  sum_comp/volume
    #print(avg_neut, avg_pos, avg_neg, avg_comp)
    
    zavg_neut = sum_neut/nonZero_neut
    zavg_pos=  sum_pos/nonZero_pos
    zavg_neg=  sum_neg/nonZero_neg
    zavg_comp=  sum_comp/nonZero_comp
    #print(zavg_neut, zavg_pos, zavg_neg, zavg_comp)

    row = {"date":date, "t_volume":volume, "t_sum_pos":sum_pos, "t_sum_neg":sum_neg, "t_sum_neut":sum_neut, "t_sum_comp":sum_comp, "t_avg_neut":avg_neut, "t_avg_pos":avg_pos, "t_avg_neg":avg_neg, "t_avg_comp":avg_comp,"t_zavg_neut":zavg_neut, "t_zavg_pos":zavg_pos, "t_zavg_neg":zavg_neg, "t_zavg_comp":zavg_comp, "t_nonZero_neut":nonZero_neut, "t_nonZero_pos":nonZero_pos, "t_nonZero_neg":nonZero_neg, "t_nonZero_comp":nonZero_comp, "t_day_of_week":0, "t_is_weekend":0}
    df = df.append(row, ignore_index=True)
    del ddf
    return df
```
