```python
def AddToPandas(globalPd, file):
    secondaryPd = pd.read_csv(file)   
    frames = [globalPd, secondaryPd]
    result = pd.concat(frames)
    print("File "+str(file)+" with shape " + str(secondaryPd.shape)+ "was added to global Df with shape"+ str(result.shape))
    return result
```
