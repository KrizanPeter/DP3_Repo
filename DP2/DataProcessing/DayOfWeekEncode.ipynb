```python
from urllib.request import urlopen, Request
from urllib.request import urlopen
from matplotlib import pyplot as plt
import math
import datetime
from time import sleep
from bs4 import BeautifulSoup
import copy
import json
import sys
import csv
import pandas as pd
import numpy as np
```

```python
def AddDayOfWeekColumn(globalPd):
    daysOfWeek_sin = []
    daysOfWeek_cos = []
    isWeekend = []
    for index, row in globalPd.iterrows():
        dayOfWeek = datetime.datetime.strptime(row['date'], '%Y-%m-%d %H:%M:%S').weekday()
        daysOfWeek_sin.append(math.sin((2*math.pi)/7*dayOfWeek))
        daysOfWeek_cos.append(math.cos((2*math.pi)/7*dayOfWeek))
        if dayOfWeek == 6 or dayOfWeek == 7:
            isWeekend.append(1)
        else:
            isWeekend.append(0)
    globalPd['dayOfWeek_sin'] = daysOfWeek_sin
    globalPd['dayOfWeek_cos'] = daysOfWeek_cos
    globalPd['isWeekend'] = isWeekend
    return globalPd
```
