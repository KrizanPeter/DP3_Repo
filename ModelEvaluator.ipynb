```python
def GetSimpleResultMetrics(y_test, y_pred):
    print(classification_report(y_test,y_pred))
    print(accuracy_score(y_test, y_pred))
    
    
def OpenOwnModelDataframe():
    df = pd.read_csv('data/OwnModel.csv')    
    return df

def OpenThreeClassOwnModelDataframe():
    df = pd.read_csv('data/3OwnModel.csv')    
    return df

def CreateOwnModelDataframe():
    df = pd.DataFrame(columns=['Model', 'Parameters', 'Accuracy', 'PrecisionRise', 'PrecisionFall', 'RecallRise', 'RecalFall', 'F1Rise', 'F1Fall'])
    return df

def SaveOwnModel(df):
    df.to_csv('data/OwnModel.csv', index=False)

def SaveThreeClassOwnModel(df):
    df.to_csv('data/3OwnModel.csv', index=False)
    
def Create2ClassResultDataframe():
    df = pd.DataFrame(columns=['Model', 'Parameters', 'Accuracy', 'PrecisionRise', 'PrecisionFall', 'RecallRise', 'RecalFall', 'F1Rise', 'F1Fall'])
    return df

def OpenTwoClassResult():
    df = pd.read_csv('data/2ClassResults.csv')    
    return df

def AppendToDf(df, model, param, acc, precR, precF, recR, recF, f1r, f1f):
    df = df.append({'Model': model, 'Parameters': param, 'Accuracy': acc, 'PrecisionRise': precR,'PrecisionFall': precF,'RecallRise': recR,'RecalFall': recF,'F1Rise': f1r, 'F1Fall': f1f }, ignore_index=True)
    return df

def CreateResultAndAppend(df, model, param, y_test, y_pred):
    report = classification_report(y_test,y_pred, output_dict=True)
    #print(classification_report(y_test,y_pred))
    
    acc = accuracy_score(y_test, y_pred)
    precR = report['Rise']['precision']
    precF = report['Fall']['precision']
    recR = report['Rise']['recall']
    recF = report['Fall']['recall']
    f1r = report['Rise']['f1-score']
    f1f = report['Fall']['f1-score']
    df = AppendToDf(df, model, param, acc, precR, precF, recR, recF, f1r, f1f)
    return df

def CreateMultipleResultAndAppend(df, model, param, y_testm, y_predm):
    samples = len(y_testm)
    acc  = 0.0
    precR = 0.0
    precF = 0.0
    recR = 0.0
    recF = 0.0
    f1r = 0.0
    f1f = 0.0
    for i in range(0,samples):
        y_test = y_testm[i]
        y_pred = y_predm[i]
        #import pdb; pdb.set_trace()
        report = classification_report(y_test,y_pred, output_dict=True)
        #print(classification_report(y_test,y_pred))

        acc += accuracy_score(y_test, y_pred)
        precR += report['Rise']['precision']
        precF += report['Fall']['precision']
        recR += report['Rise']['recall']
        recF += report['Fall']['recall']
        f1r += report['Rise']['f1-score']
        f1f += report['Fall']['f1-score']
        
    df = AppendToDf(df, model, param, acc/samples, precR/samples, precF/samples, recR/samples, recF/samples, f1r/samples, f1f/samples)
    return df


def SaveTwoClassResult(df):
    df.to_csv('data/2ClassResults.csv', index=False)
```

```python
def Create3ClassResultDataframe():
    df = pd.DataFrame(columns=['Model', 'Parameters', 'Accuracy', 'PrecisionRise', 'PrecisionFall','PrecisionStay', 'RecallRise', 'RecalFall','RecalStay', 'F1Rise', 'F1Fall', 'F1Stay'])
    return df

def OpenThreeClassResult():
    df = pd.read_csv('data/3ClassResults.csv')    
    return df

def AppendToThreeDf(df, model, param, acc, precR, precF, precS, recR, recF, recS, f1r, f1f, f1s):
    df = df.append({'Model': model, 'Parameters': param, 'Accuracy': acc, 'PrecisionRise': precR,'PrecisionFall': precF, 'PrecisionStay': precS ,'RecallRise': recR,'RecalFall': recF,'RecalStay':recS,'F1Rise': f1r, 'F1Fall': f1f, 'F1Stay': f1s }, ignore_index=True)
    return df

def CreateResultAndAppendThreeClass(df, model, param, y_test, y_pred):
    report = classification_report(y_test,y_pred, output_dict=True)
    #print(classification_report(y_test,y_pred))
    
    acc = accuracy_score(y_test, y_pred)
    precR = report['Rise']['precision']
    precF = report['Fall']['precision']
    precS = report['Stay']['precision']
    recR = report['Rise']['recall']
    recF = report['Fall']['recall']
    recS = report['Stay']['recall']
    f1r = report['Rise']['f1-score']
    f1f = report['Fall']['f1-score']
    f1s = report['Stay']['f1-score']
    df = AppendToThreeDf(df, model, param, acc, precR, precF, precS, recR, recF, recS, f1r, f1f, f1s)
    return df

def CreateMultipleResultAndAppendThreeClass(df, model, param, y_testm, y_predm):
    samples = len(y_testm)
    acc = 0.0
    precR =  0.0
    precF = 0.0
    precS  = 0.0
    recR  = 0.0
    recF = 0.0
    recS = 0.0
    f1r = 0.0
    f1f = 0.0
    f1s = 0.0
    
    for i in range(0,samples):
        y_test = y_testm[i]
        y_pred = y_predm[i]
        report = classification_report(y_test,y_pred, output_dict=True)  
        acc += accuracy_score(y_test, y_pred)
        precR += report['Rise']['precision']
        precF += report['Fall']['precision']
        precS += report['Stay']['precision']
        recR += report['Rise']['recall']
        recF += report['Fall']['recall']
        recS += report['Stay']['recall']
        f1r += report['Rise']['f1-score']
        f1f += report['Fall']['f1-score']
        f1s += report['Stay']['f1-score']
        #import pdb; pdb.set_trace()
    df = AppendToThreeDf(df, model, param, acc/samples, precR/samples, precF/samples, precS/samples, recR/samples, recF/samples, recS/samples, f1r/samples, f1f/samples, f1s/samples)
    return df

def SaveThreeClassResult(df):
    df.to_csv('data/3ClassResults.csv', index=False)
```
